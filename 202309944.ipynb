```python
import pandas as pd
import os

# Heap 클래스
class Heap:
    def __init__(self):
        self.data = []

    def insert(self, item):
        self.data.append(item)
        self.data.sort()  # 간단 구현

    def deleteMax(self):
        if len(self.data) == 0:
            return None
        return self.data.pop()

# Create a dummy birthday.csv file if it doesn't exist
file_name = "birthday.csv"
if not os.path.exists(file_name):
    dummy_data = {
        "Unnamed: 0": ["Alice", "Bob", "Charlie", "David", "Eve"],
        "생년(예. 1985)": [1990, 1988, 1992, 1985, 1995],
        "생월(예. 12)": [1, 5, 10, 3, 7],
        "생일(예. 5)": [15, 20, 25, 10, 30]
    }
    dummy_df = pd.DataFrame(dummy_data)
    dummy_df.to_csv(file_name, index=False)
    print(f"Dummy '{file_name}' created.")

# CSV 읽기
df = pd.read_csv("birthday.csv")

h = Heap()

# 힙에 데이터 넣기
for _, row in df.iterrows():
    try:
        name = row["Unnamed: 0"]
        y = int(row["생년(예. 1985)"])
        m = int(row["생월(예. 12)"])
        d = int(row["생일(예. 5)"])

        key = y*10000 + m*100 + d
        h.insert((key, name, y, m, d))

    except Exception as e:
        print(f"Error processing row: {e}, {row}")
        continue

# 결과 출력
print("Top 10 birthdays (latest first):")
for i in range(10):
    data = h.deleteMax()
    if data:
        key, name, y, m, d = data
        print(f"{name} {y} {m} {d}")
```

## 5. 우선순위 큐 연습문제 풀이 (01~07)


### 01
**가질 수 있다.**  
최대 힙의 조건은 **부모의 키가 자식의 키보다 크거나 같다**는 것뿐이다.  
따라서 더 얕은 깊이에 있는 어떤 노드가 더 깊은 곳의 어떤 노드보다 작을 수도 있다.  
즉, 힙은 **전체 정렬 구조가 아니라 부분 순서 구조**이다.

### 02
**항상 가장 작은 값은 아니다.**  
최대 힙에서 `A[0]`이 최댓값인 것은 맞지만, 마지막 원소 `A[n-1]`은 단지 **마지막 레벨의 맨 끝 노드**일 뿐이다.  
최솟값은 보통 리프들 중 하나에 있지만, 반드시 마지막 원소일 필요는 없다.

### 03
`buildHeap()`에서 스며내리기를 **검사조차 하지 않고 넘어가는 노드 수**는 정확히 **리프 노드 수**이다.  
배열 기반 힙에서 인덱스 `⌊n/2⌋` 이상인 노드는 모두 리프이므로 그 개수는

\[
n - \lfloor n/2 \rfloor = \lceil n/2 \rceil
\]

이다.

### 04
배열 `A[0...n-1]`에 대해 **한 번의 스며내리기(sift-down)** 수행 시간은 다음과 같다.

- **최악의 경우:** 루트에서 리프 가까이까지 내려가야 하므로 `Θ(log n)`
- **최선의 경우:** 부모가 이미 자식들보다 커서 바로 종료되므로 `Θ(1)`

### 05
**간단한 일이다.**  
힙의 맨 마지막 원소를 삭제하는 것은 힙 순서를 깨뜨리지 않는다.  
왜냐하면 마지막 원소는 가장 아래, 가장 오른쪽에 있으므로 그것을 제거해도 **완전 이진 트리 조건**이 유지되고,  
부모-자식 관계도 새로 위반되지 않기 때문이다.  
따라서 마지막 원소 삭제는 보통 **배열의 마지막 원소를 제거하는 것**과 같아서 `Θ(1)`이다.

### 06
위에서부터 시작해서 각 원소를 **스며오르기(sift-up)** 시키는 방법으로도 힙은 만들 수 있다.  
예를 들어 배열의 앞에서부터 원소를 하나씩 힙에 삽입한다고 생각하면 된다.

하지만 효율은 더 나쁘다.

- **본문의 `buildHeap()`**: 아래에서부터 스며내리기 → `Θ(n)`
- **위에서부터 스며오르기 반복**: 각 삽입마다 최대 `Θ(log n)` → 전체 `Θ(n log n)`

따라서 **본문 방법이 점근적으로 더 효율적이다.**

### 07
최대 힙에서 어떤 원소의 값이 증가하면 그 원소가 **부모보다 커질 가능성**이 생긴다.  
이때는 그 원소를 **위로 끌어올리는 스며오르기(sift-up)** 를 수행하면 된다.

과정은 다음과 같다.

1. 값이 증가한 원소의 위치를 찾는다.
2. 그 원소와 부모를 비교한다.
3. 부모보다 크면 서로 교환한다.
4. 더 이상 부모보다 크지 않거나 루트에 도달할 때까지 반복한다.

힙의 높이는 `O(log n)`이므로 이 수선 과정 전체 시간은 **`O(log n)`** 이다.
