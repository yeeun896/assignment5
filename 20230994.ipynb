```python
import pandas as pd
import os

# Heap 클래스
class Heap:
    def __init__(self):
        self.data = []

    def insert(self, item):
        self.data.append(item)
        self.data.sort()  # 간단 구현

    def deleteMax(self):
        if len(self.data) == 0:
            return None
        return self.data.pop()

# Create a dummy birthday.csv file if it doesn't exist
file_name = "birthday.csv"
if not os.path.exists(file_name):
    dummy_data = {
        "Unnamed: 0": ["Alice", "Bob", "Charlie", "David", "Eve"],
        "생년(예. 1985)": [1990, 1988, 1992, 1985, 1995],
        "생월(예. 12)": [1, 5, 10, 3, 7],
        "생일(예. 5)": [15, 20, 25, 10, 30]
    }
    dummy_df = pd.DataFrame(dummy_data)
    dummy_df.to_csv(file_name, index=False)
    print(f"Dummy '{file_name}' created.")

# CSV 읽기
df = pd.read_csv("birthday.csv")

h = Heap()

# 힙에 데이터 넣기
for _, row in df.iterrows():
    try:
        name = row["Unnamed: 0"]
        y = int(row["생년(예. 1985)"])
        m = int(row["생월(예. 12)"])
        d = int(row["생일(예. 5)"])

        key = y*10000 + m*100 + d
        h.insert((key, name, y, m, d))

    except Exception as e:
        print(f"Error processing row: {e}, {row}")
        continue

# 결과 출력
print("Top 10 birthdays (latest first):")
for i in range(10):
    data = h.deleteMax()
    if data:
        key, name, y, m, d = data
        print(f"{name} {y} {m} {d}")
```

```python
import pandas as pd
# from DS.list.circularDoublyLinkedList import *

# Dummy CircularDoublyLinkedList class to resolve ModuleNotFoundError
class CircularDoublyLinkedList:
    def __init__(self):
        self.nodes = []

    def append(self, data):
        self.nodes.append(data)

    def printList(self):
        if not self.nodes:
            print("List is empty.")
            return
        print("CircularDoublyLinkedList contents:")
        for node_data in self.nodes:
            print(node_data)

# CSV 읽기
df = pd.read_csv("birthday.csv")

# 원형 이중 연결 리스트 생성
friends = CircularDoublyLinkedList()

# 같은 조 친구들 저장
for _, row in df.iterrows():
    try:
        name = row["Unnamed: 0"]
        year = row["생년(예. 1985)"]
        month = row["생월(예. 12)"]
        day = row["생일(예. 5)"]

        # 생년월일 누락 예외처리
        if pd.isna(year) or pd.isna(month) or pd.isna(day):
            raise ValueError("생년월일 없음")

        info = f"{name}: {int(year)}-{int(month):02d}-{int(day):02d}"
        friends.append(info)

    except Exception:
        friends.append(f"{row['Unnamed: 0']}: 생년월일 정보 없음")

# 출력
friends.printList()
```
